```python
import pandas as pd
```

```python
num_min_in_period = 20

full_df = pd.DataFrame(columns=["n", "k", "n_choose_k"])
full_df.to_pickle("full_df.pkl")

for i in range(0, num_min_in_period):
    # Load partial df
    n_start = i*60
    n_end = (i+1)*60
    partial_df_path = f"partial_dfs/df_{str(n_start)}-{str(n_end)}.pkl"
    partial_df = pd.read_pickle(partial_df_path)
    if i != num_min_in_period - 1:
        # Leave out the last minute of data so we don't duplicate
        partial_df = partial_df[partial_df["n"] != n_end]

    # Combine with full df
    full_df = pd.concat([full_df, partial_df], ignore_index=True)

display(full_df)
full_df.to_pickle("precomputed_combinations_df.pkl")
```
